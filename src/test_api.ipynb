```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()
token = os.getenv("BEARER_TOKEN")
url = os.getenv("CREWAI_URL")
```

```python
headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "Application/json" 
}
```

```python
response = requests.get(f"{url}/inputs", headers=headers)
if response.status_code == 200:
    print(response.json())
else:
    print(f"Erro {response.status_code}")
```

```python
body = {
    "inputs": {
        "topic": "CrewAI",
        "current_year": 2026
    }
}

response = requests.post(f"{url}/kickoff", headers=headers, json=body)
if response.status_code == 200:
    print(response.json())
else:
    print(f"Erro {response.status_code}")
```

```python
kickoff_id = "79f96022-52e1-43ee-947a-87b9ef34ee5b"
response = requests.get(f"{url}/status/{kickoff_id}", headers=headers)
if response.status_code == 200:
    print(response.json())
else:
    print(f"Erro {response.status_code}")
```
